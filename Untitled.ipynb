1. X坐标应沿着水平轴放置。
2. Y坐标应该沿着垂直轴放置。
3. 所有的轨迹都应该显示在同一张图上
4. 所有轨迹都应该是可区分的（通过颜色，线条类型，标记-适合就行）。
5. 图形应该具有标题"Визуализация дорожной сцены"。
6. 轴的名称应在图表上注明.
7. 图形上应显示一个网格。
8. 图例应显示图形上的轨迹及其来自文件的标识符（TRACK_ID）和对象类型（OBJ_TYPE）的对应关系。
9. 为了简化，表单00000000-0000-0000-0000000ABCDF的标识符可以替换为00000000ABCDF（OBJ_TYPE）。

```python
import matplotlib.pyplot as plt
import numpy as np
import csv
```

```python
filename = "data.csv"
with open(filename) as file:
    records = csv.reader(file)
    # 去除表头
    next(records)
    #
    record_dict = {}  #  {"track_id": [{timestamp: (x, y)}, ...]}
    for record in records:
        timestamp = float(record[0])
        track_id = record[1]
        object_type = record[2]
        x = float(record[3])
        y = float(record[4])
        city_name = record[5]
        track_id = "{}-{}".format(track_id, object_type)
        # print(timestamp, track_id, object_type, x, y, city_name)
        if record_dict.get(track_id) is None:
            record_dict[track_id] = [{timestamp: (x, y)}]
        else:
            record_dict[track_id].append({timestamp: (x, y)})
record_dict
```

```python
#创建figure窗口
fig = plt.figure(num=3, figsize=(10, 10))

plt.title("Визуализация дорожной сцены",fontdict = {'color':'black','size':15})

plt.grid()  # 生成网格

plt.xlabel('X')
plt.ylabel('Y')

# 范围
min_x = 1909.5366475680846
min_y = 506.6113991835553
max_x = 1999.1508266467456
max_y = 646.3402280796662

# plt.xlim((1640, 1770))
# plt.ylim((370, 510))
plt.xlim((1880, 2010))
plt.ylim((480, 660))
#设置坐标轴刻度
# my_x_ticks = np.arange(1640, 1780, 20)
# my_y_ticks = np.arange(380, 520, 20)
my_x_ticks = np.arange(1880, 2020, 20)
my_y_ticks = np.arange(490, 670, 20)
plt.xticks(my_x_ticks)
plt.yticks(my_y_ticks)

# 画点
labels = []
colormap = plt.cm.gist_ncar #nipy_spectral, Set1,Paired  
colorst = [colormap(i) for i in np.linspace(0, 1, len(record_dict) + 2)] 
index = 0
for track_id, records in record_dict.items():
#     print(track_id)
    for record in records:
#         print(record)
        for timestram, point in record.items():
            x, y = point
#             print(track_id)
            track_tag = track_id.split("-")[-2]
            object_type = track_id.split("-")[-1]
            label = "{} ({})".format(track_tag, object_type)
            if label in labels:
                plt.plot(x, y, '+', c=colorst[index])
            else:
                plt.plot(x, y, '+', c=colorst[index], label=label, linestyle='-')
                labels.append(label)
    index +=1 

plt.legend(loc="upper left")


# ax = fig.add_subplot(111)
# ax.grid()
# for x,y,lab in zip(X,Y,labels):
#         ax.scatter(x,y,label=lab)
        
#Now this is actually the code that you need, an easy fix your colors just cut and paste not you need ax.
# colormap = plt.cm.gist_ncar # nipy_spectral, Set1,Paired  
# colorst = [colormap(i) for i in np.linspace(0, 0.9,len(ax.collections))]       
# for t,j1 in enumerate(ax.collections):
#     j1.set_color(colorst[t])

plt.show()
```
